**Null values:** 
- missing values in a dataset 

Remedy: 
    - if the **missing values** are more than 80% - drop the column (delete or exclude the column from your analysis)
    - if the **missing values** are less than 20% - we drop the rows containing the missing values
    - if the **missing values** are above 40% or 50% -  we conduct what we call **imputation** (we fill in the missing values using eg an MCT)

#### **Measures of Central Tendency**
- Attaining one numerical value that is represtative of the entire data set( feature).  
     - MEAN(arithmetic mean)
     - MEDIAN(after arraing your data either in ascending or descending order you get the center item) 
     - MODE(most frequently repeated value)
     

```python
mean = df["Cement (component 1)(kg in a m^3 mixture)"].mean()
mean
```

#### **Univariate Analysis**
- analysis of single features

```python
df["Cement (component 1)(kg in a m^3 mixture)"].describe()
```

```python
IQR = df["Cement (component 1)(kg in a m^3 mixture)"].quantile(0.75) - df["Cement (component 1)(kg in a m^3 mixture)"].quantile(0.25) 
IQR
```

```python
df_range = df["Cement (component 1)(kg in a m^3 mixture)"].max() - df["Cement (component 1)(kg in a m^3 mixture)"].min()
df_range
```

```python
df_variance = df["Cement (component 1)(kg in a m^3 mixture)"].var()
df_variance
```

```python
df_sd = df["Cement (component 1)(kg in a m^3 mixture)"].std()
df_sd
```

```python
df_median = df["Cement (component 1)(kg in a m^3 mixture)"].median()
df_median
```

```python
df_mode = df["Cement (component 1)(kg in a m^3 mixture)"].mode()
df_mode
```
